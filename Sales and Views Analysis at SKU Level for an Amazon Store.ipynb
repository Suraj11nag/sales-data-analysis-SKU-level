```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
sales_data = pd.read_csv('Sales_Views.csv')
glance_views = pd.read_csv('Glance_Views.csv')
```

```python
sales_data = sales_data.drop(columns=['Unnamed: 0','Unnamed: 10'])
sales_data
```

```python
glance_views = glance_views.drop(columns=['Unnamed: 0'])
glance_views
```

```python
glance_views.info()
```

```python
sales_data.info()
```

```python
df = pd.merge(sales_data, glance_views, on=['SKU_NAME'], how='right')
```

```python
df
```

```python
sum(df[df['SKU_NAME']=='C120[H:8NV'].value_counts() == 1)
```

```python
df['avg_selling_price'] = df['ORDERED_REVENUE'] / df['ORDERED_UNITS']
```

```python
df['avg_selling_price']
```

```python
df['unit_conversion'] = df['ORDERED_UNITS'] / df['PAGE VIEWS']
```

```python
df['unit_conversion']
```

```python
df.head()
```

```python
subset = df.loc[df['SKU_NAME'] == 'C120[H:8NV', ['unit_conversion', 'avg_selling_price']]
if len(subset) >= 2:
    corr, p_value = pearsonr(subset['unit_conversion'], subset['avg_selling_price'])
    print("Correlation coefficient:", corr)
    print("P-value:", p_value)
else:
    print("Insufficient data to compute correlation coefficient.")
```
